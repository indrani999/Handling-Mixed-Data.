# Working with Date

```python
# Converting the data type of date column
```

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1. Extract Year

```python
date['date_year'] = date['date'].dt.year #This will display the year in that particular column
date.head()
```

2. Extract Month

```python
date['date_month_no'] = date['date'].dt.month #This will display the month number in that particular column
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name() #This will display the month name in that particular column
date.head()
```

3. Extract days

```python
date['date_day'] = date['date'].dt.day #This will display the date in that particular column
date.head()
```

```python
# Day of the week
```

```python
date['date_dow'] = date['date'].dt.dayofweek #This will display the day of the week
date.head()
```

```python
# Day Name of the week
```

```python
date['date_dow_name'] = date['date'].dt.day_name() #This will display the day of the week
date.head()
```

```python
# If weekends?
```

```python
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0) #This will display the day of the week
date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract week of the year

```python
date['date_week'] = date['date'].dt.week
date.head()
```

# Extract Quarter

```python
date['Quarter'] = date['date'].dt.quarter
date.head()
```

# Extract Semester

```python
date['semester'] = np.where(date['Quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```
